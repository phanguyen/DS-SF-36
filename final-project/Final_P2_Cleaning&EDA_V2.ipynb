# Part 1: Prepare & Cleaning Data

### V1: contain only mass_shooting data of 2016 & 2017 (several months)

```python
import os

import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 20)
pd.set_option('display.notebook_repr_html', True)
pd.set_option('display.max_columns', 12)
```

```python
#read each csv file with pd
Mass_df1 = pd.read_csv(os.path.join('Mass1.csv'))
Mass_df2 = pd.read_csv(os.path.join('Mass2.csv'))
Mass_df3 = pd.read_csv(os.path.join('Mass3.csv'))
Mass_df4 = pd.read_csv(os.path.join('Mass4.csv'))
Mass_df5 = pd.read_csv(os.path.join('Mass5.csv'))
Mass_df6 = pd.read_csv(os.path.join('Mass6.csv'))
Mass_df7 = pd.read_csv(os.path.join('Mass7.csv'))
Mass_df8 = pd.read_csv(os.path.join('Mass7.csv'))
Mass_df9 = pd.read_csv(os.path.join('Mass7.csv'))
Home_df1 = pd.read_csv(os.path.join('HomeInvasion1.csv'))
Home_df2 = pd.read_csv(os.path.join('HomeInvasion2.csv'))
Home_df3 = pd.read_csv(os.path.join('HomeInvasion3.csv'))
Home_df4 = pd.read_csv(os.path.join('HomeInvasion4.csv'))
Home_df5 = pd.read_csv(os.path.join('HomeInvasion5.csv'))
```

```python
Mass_List = [Mass_df1, Mass_df2, Mass_df3,Mass_df4,Mass_df5,Mass_df6,Mass_df7,Mass_df8,Mass_df9]
Mass_df = pd.concat(Mass_List)

Home_List = [Home_df1,Home_df2,Home_df3]
Home_df = pd.concat(Home_List)
```

```python
type(Mass_df)
```

```python
Mass_df['Incident'] = 'Mass_Shooting'
```

```python
Home_df['Incident'] = 'Home_Invasion'
```

```python
Mass_Home = [Mass_df,Home_df]
Mass_Home_df = pd.concat(Mass_Home)
Mass_Home_df
```

```python
Law_df = pd.read_csv(os.path.join('Gun_Law.csv'))
Campus_df = pd.read_csv(os.path.join('Gun_On_Campus.csv'))
```

```python
Law_And_Campus_df = pd.merge(Law_df, Campus_df, on="State")
Law_And_Campus_df
```

```python
df = pd.merge(Mass_Home_df, Law_And_Campus_df, on="State")
df
```
